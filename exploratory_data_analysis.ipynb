```python
import json
import numpy
import scipy
import pandas
import matplotlib
import sklearn
import os
from collections import Counter
```

```python
cooking_data = json.load(open("train.json"))
```

```python
# Cuisine Exploratory Analysis

def list_cuisine_all(data): # list of all cuisines
    cuisine_list = []
    for name in data:
        cuisine_list.append(name['cuisine'])
    return cuisine_list
    
def cuisine_count(data): # counter object of unique cuisines with count  
    cuisine_list = list_cuisine_all(data)
    c = Counter(cuisine_list)
    return c
```

```python
#Ingredient Exploratory Analysis

def list_ingredients_all(data): # list of all ingredients
    ingredients_list = []
    for name in data:
        ingredient_list = name['ingredients']
        for ingred in ingredient_list:
            ingredients_list.append(ingred)
    return ingredients_list

def list_ingredients(data): #list of unique ingredients
    ingredients_set = set()
    for name in data:
        ingredient_list = name['ingredients']
        for ingred in ingredient_list:
            ingredients_set.add(ingred)
    return ingredients_set
```

```python
print("Cuisine Analysis")
print("\n")
print ("There are a total of", len(list_cuisine_all(cooking_data)), "recipes.")
print ("Total number of recipes for each cuisine:", cuisine_count(cooking_data))
print ("There are ", len(cuisine_count(cooking_data)), "unique cuisines.")
print("\n")      
print("Ingredient Analysis")
print("\n")
print ("There are a total of", len(list_ingredients_all(cooking_data)), "ingredients.")
print ("Total number of recipes for each ingredient: ", ingredient_count(cooking_data))
print ("There are ", len(list_ingredients(cooking_data)), "unique ingredients.")
```
